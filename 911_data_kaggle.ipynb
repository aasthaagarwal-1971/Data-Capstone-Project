```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
#reading the data
df = pd.read_csv('911.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['zip'].value_counts().head()
```

```python
df['twp'].value_counts().head(5)
```

```python
df['title'].nunique()
```

```python
#but the title majorily lies in three categories, EMS,Fire and traffic, so we make a new column for that to segregate better
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
df['Reason'].value_counts()
```

```python
sns.countplot(x='Reason',data=df,palette='viridis')
#this emplies, EMS is the biggest reason
```

```python
#as the the data in time column is in str form, we need to convert it to date time dtype
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
#the analysis hourly and monthy wise
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
sns.countplot(x='Day of Week',data=df,hue='Reason')
```

```python
sns.countplot(x='Month',data=df,hue='Reason')
#months 9,10,11 are missing
```

```python
byMonth = df.groupby('Month').count()
byMonth.head(9)
```

```python
#possibly a simple line plot that fills in the missing months
byMonth['twp'].plot()
```

```python
sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```

```python
#creating a column having date of the timestamp
df['Date']=df['timeStamp'].apply(lambda t: t.date())
df.head()
```

```python
#the number of counts of 911 calls date-wise
df.groupby('Date').count()['twp'].plot()
plt.tight_layout()
```

```python
#plots reason wise
df[df['Reason']=='Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['twp'].plot()
plt.title('EMS')
plt.tight_layout()
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
dayHour.head()
```

```python
#we can get a better idea via heatmaps now
#hour as column
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='viridis')
```

```python
#months as column
dayMonth = df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
dayMonth.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth,cmap='viridis')
```
